single Layer Perceptron for OR gate for Bipolar Inputs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Input = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
Target = np.array([[-1], [1], [1], [-1]])

W1 = 0
w2 = 0
b = 0
Learning_rate = 0.3
thereshold = 1
epochs = 10
```

```python
def activation(net):
    return 1 if net > thereshold else -1
```

```python
for epoch in range(epochs):
    error = 0
    for i in range(len(Input)):
        net = Input[i][0]*W1 + Input[i][1]*w2 + b
        output = activation(net)

        loss = Target[i] - output
        error += abs(loss)

        if loss != 0:
            W1 += Learning_rate * error * Input[i][0]
            w2 += Learning_rate * error * Input[i][1]
            b += Learning_rate * error

    print(f'Epoch {epoch+1} : Error = {error}')

    if error == 0:
        break
    
    print("Final Weights :")
    print(f'W1 = {W1}')
    print(f'w2 = {w2}')
    print(f'b = {b}')

    for i in range(len(Input)):
        net = Input[i][0]*W1 + Input[i][1]*w2 + b
        output = activation(net)
        print(f'Output for {Input[i]} = {output}')    
```

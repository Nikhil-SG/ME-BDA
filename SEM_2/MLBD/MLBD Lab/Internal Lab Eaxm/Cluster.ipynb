k-means clustering and agglomerative hierarchical clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('Country-data.csv')
X = data.iloc[:, 1:].values
```

```python
# KMeans
wcss = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,10), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# KMeans
kmeans = KMeans(n_clusters=2, max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)
data['Cluster'] = y_kmeans
```

```python
# Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=2)
y_agglomerative = agglomerative.fit_predict(X)
data['Cluster2'] = y_agglomerative
```

```python
# compare the two clustering methods by plotting the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.title('KMeans')
plt.show()

plt.scatter(X[y_agglomerative == 0, 0], X[y_agglomerative == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(X[y_agglomerative == 1, 0], X[y_agglomerative == 1, 1], s=100, c='blue', label='Cluster 2')
plt.title('Agglomerative Clustering')
plt.show()
```

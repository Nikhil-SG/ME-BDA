```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('RDD').getOrCreate()
```

```python
type(spark)
```

```python
Bank = spark.sparkContext.textFile('/content/drive/MyDrive/BDA_1_Sem/ABD/ABD_Lab/Hadoop/RDD Lab assignments/bank.txt')
Bank.collect()
```

Q1) Count unique number of customers

```python
cust = Bank.map(lambda x:x.split('\t'))
cust.collect()
```

```python
unicust = cust.map(lambda x:x[1]).distinct().count()
unicust
```

Q2) Count unique number of Bank ID


```python
uninum=cust.map(lambda x:x[0]).distinct().count()
uninum
```

Q3) Count unique number of customers per Bank ID

```python
unibankcust = cust.map(lambda x:(x[0],x[1])).distinct().map(lambda d:(d[0],1)).reduceByKey(lambda x,y:x+y)
unibankcust.collect()
```

Q4) Number of transactions for given Account Number

```python
acc_no='123872829'
broadcast_account_number=spark.sparkContext.broadcast(acc_no)
print(broadcast_account_number.value)

transactions_for_account = cust.filter(lambda x: x[1] == broadcast_account_number.value)
print(transactions_for_account.collect())

num_transaction=transactions_for_account.count()

print(f"Number of transactions for account {acc_no}: {num_transaction}")
```

Q5) Number of credit transactions for given Account Number in a given year

```python
bank_year=cust.map(lambda x:(x[1],x[3],x[2].split('-')[2]))
bank_year.collect()
```

```python
bank_year.filter(lambda x:(x[0]=='123872829') & (x[1]=='C') & (x[2]=='2001')).count()
```

```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('RDD').getOrCreate()
```

Q1) Compute number of students from each Institute.

```python
MAHE = spark.sparkContext.textFile("/content/drive/MyDrive/BDA_1_Sem/ABD/ABD_Lab/Hadoop/RDD Lab assignments/student.txt")
student = MAHE.map(lambda x:x.split(' '))
MAHE.collect()
```

```python
institue=student.map(lambda x:(x[1],1)).reduceByKey(lambda x,y:x+y)
institue.collect()
```

Q2) Number of students enrolled to any program.

```python
program_count=student.map(lambda x:(x[2],1)).reduceByKey(lambda x,y:x+y)
program_count.collect()
```

Q3) Number of ‘boy’ and ‘girl’ students

```python
Gender_count=student.map(lambda x:(x[3],1)).reduceByKey(lambda x,y:x+y)
Gender_count.collect()
```

Q4) Number of ‘boy’ and ‘girl’ students from selected Institute.

```python
selected_institue=student.filter(lambda x:x[1]=='MSIS')
selected_institue.collect()
```

```python
Gender_in_MSIS=selected_institue.map(lambda x:(x[3],1)).reduceByKey(lambda x,y:x+y)
Gender_in_MSIS.collect()
```

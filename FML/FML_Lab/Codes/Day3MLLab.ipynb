```python
Hour = int(input("Enter Hours: "))
Rate = float(input("Enter The rate : "))
try:
  Pay = Hour * Rate
  print("Pay :",Pay)
except:
  Hour > 40
  Pay = (Hour - 40) * (1.5 * Rate) + (40 * Rate)
  print("Pay :",Pay)
```

Python Program to Exchange the Values of Two Numbers Without Using a Temporary Variable.

```python
def swapnumber(a,b):
  a = a + b
  b = a - b
  a = a - b
  return a,b

if __name__ == "__main__":
  a = int(input("Enter the number: "))
  b = int(input("Enter the number: "))
  a,b = swapnumber(a,b)
  print("The numbers after swapping are",a,b)


```

Python Program to Print all Numbers in a Range Divisible by a Given Number. [ if range is from 1
to 20 and number is 4, then the result should be 4, 8, 12, 16 and 20. ]


```python
start = int(input("Enter the start range of the number: "))
end = (int(input("Enter the end range of the number: ")))
num = int(input("Enter the number: " ))
count = 0
divisible_numbers= []
for i in range(start, end+1):
  if i % num == 0 :
    count += 1
    divisible_numbers.append(i)
print("In Range of numbers given from",start,"to",end,"are divide",count,"times by the number",num,". They are",divisible_numbers)
```

Python Program to Print Odd Numbers Within a Given Range.


```python
num = int(input("Enter the range of number: "))
print("The odd numbers in the given range are: ")
count = 0
oddnum = []
for i in range(1, num + 1):
  if i % 2 != 0:  # Check if 'i' is odd
    count += 1
    oddnum.append(i)
print(oddnum, "the number of odd numbers present in the range are", count)
```

Python Program to Find the Sum of Digits in a Number.

```python
def sumofdigit(num):
  sum = 0
  while num > 0:
    digit = num % 10
    sum += digit
    num = num // 10
  return sum

if __name__ == "__main__":
  num = int(input("Enter the number: "))
  sum = sumofdigit(num)
  print("The sum of digits of the number",num,"is",sum)
```

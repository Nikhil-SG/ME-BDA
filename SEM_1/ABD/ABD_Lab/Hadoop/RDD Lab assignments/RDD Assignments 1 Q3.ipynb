Q1) Find maximum and minimum temperature of all cities from the given dataset

```python
temp = spark.sparkContext.textFile('/content/drive/MyDrive/BDA_1_Sem/ABD/ABD_Lab/Hadoop/RDD Lab assignments/sampleTemp.txt')
temp.collect()
```

```python
data=temp.map(lambda x:x.split(' '))
data.collect()
```

```python
city_temp=data.map(lambda x:(x[2],x[1]),1)

max_temp=city_temp.reduceByKey(lambda x,y:max(x,y))
print(f'The maximum temperature of city are :{max_temp.collect()}')

min_temp= city_temp.reduceByKey(lambda x, y: min(x, y))
print(f'The minimum temperature of city are :{min_temp.collect()}')
```

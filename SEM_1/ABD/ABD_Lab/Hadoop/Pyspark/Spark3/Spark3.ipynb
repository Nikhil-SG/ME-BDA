```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Weather').getOrCreate()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import types as T
from pyspark.sql.window import Window
```

```python
df = spark.read.csv('/content/drive/MyDrive/BDA_1_Sem/ABD/ABD_Lab/Hadoop/Pyspark/Spark3/Weather Data in India from 1901 to 2017.csv', header = True, inferSchema = True)
df.show()
```

```python
df.printSchema()
```

Q1) Add column average temperature for each year

```python
df1 = df.withColumn('Average Temperature', (df['Jan'] + df['Feb'] + df['Mar'] + df['Apr'] + df['May'] + df['Jun'] + df['Jul'] + df['Aug'] + df['Sep'] + df['Oct'] + df['NOV'] + df['Dec'])/12)
df1.show()
```

Q2) Add row average temperature of each month

```python
avg_temperatures = df.select([avg(col(month)).alias(month) for month in df.columns[2:]]).first()

schema = T.StructType([
    T.StructField("_c0", T.IntegerType(), True),
    T.StructField("YEAR", T.IntegerType(), True),
    T.StructField("JAN", T.DoubleType(), True),
    T.StructField("FEB", T.DoubleType(), True),
    T.StructField("MAR", T.DoubleType(), True),
    T.StructField("APR", T.DoubleType(), True),
    T.StructField("MAY", T.DoubleType(), True),
    T.StructField("JUN", T.DoubleType(), True),
    T.StructField("JUL", T.DoubleType(), True),
    T.StructField("AUG", T.DoubleType(), True),
    T.StructField("SEP", T.DoubleType(), True),
    T.StructField("OCT", T.DoubleType(), True),
    T.StructField("NOV", T.DoubleType(), True),
    T.StructField("DEC", T.DoubleType(), True)
])

new_row_data = [(None, None, avg_temperatures[0], avg_temperatures[1],
                 avg_temperatures[2], avg_temperatures[3], avg_temperatures[4],
                 avg_temperatures[5], avg_temperatures[6], avg_temperatures[7],
                 avg_temperatures[8], avg_temperatures[9], avg_temperatures[10],
                 avg_temperatures[11])]

new_row_df = spark.createDataFrame(new_row_data, schema=schema)
df_combined = new_row_df.union(df)
df_combined = df_combined.orderBy("YEAR")
df_combined.show()
```

Q3) Add columns Min and Max for each year with corresponding values

```python
df3 = df1.withColumn("Min", least(*[col(month) for month in df.columns[2:13]]))
df3 = df3.withColumn("Max", greatest(*[col(month) for month in df.columns[2:13]]))
df3.show()
```

Q4) Create dataframe called 'decade', storing average temperatures of each month in the respective
decade. Decade 1910 is from 1901 to 1910.

```python
df = df.withColumn("Decade", (F.col("YEAR") / 10).cast("integer") * 10)
df.show()
decade_avg = df.groupBy("Decade").agg(*[F.avg(month).alias(month) for month in df.columns[2:]] ).orderBy("Decade", ascending = True)
decade_avg.show()
```

Q5) Which year is the hottest year? (Based on avg temp)

```python
hottest_year = df1.orderBy(F.desc("Average Temperature")).first()
coldest_year = df1.orderBy("Average Temperature").first()
print(f"Hottest Year: {hottest_year['YEAR']} with Average Temperature: {hottest_year['Average Temperature']:.2f}")
print(f"Coldest Year: {coldest_year['YEAR']} with Average Temperature: {coldest_year['Average Temperature']:.2f}")
```

Q6) Find out which years recorded minimum and maximum temperatures.

```python
hottest_year = df3.orderBy(F.desc("Max")).first()
coldest_year = df3.orderBy("Min").first()
print(f"Hottest Year: {hottest_year['YEAR']} with Max Temperature: {hottest_year['Max']:.2f}")
print(f"Coldest Year: {coldest_year['YEAR']} with Min Temperature: {coldest_year['Min']:.2f}")
```

Q7) How much raise is observed in min and max temperature for each month?

```python
df7 = df3.withColumn("Min_Difference", F.col("Min") - F.col("Average Temperature")) \
            .withColumn("Max_Difference", F.col("Max") - F.col("Average Temperature"))
df7_ = df7.select(
    "YEAR",
    "Min",
    "Max",
    "Average Temperature",
    "Min_Difference",
    "Max_Difference"
)
df7_.show()
```

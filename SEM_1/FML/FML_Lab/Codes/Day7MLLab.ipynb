```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/BDA_1_Sem/FML/FML_Lab/Datasets/Social_Network_Ads.csv')
dataset = data
dataset.head()
```

```python
X = dataset.iloc[:, [2, 3]].values
X
```

```python
Y = dataset.iloc[:, 4].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
X_train
len(X_train)
```

```python
Y_test
len(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, Y_train)
```

```python
Y_pred = classifier.predict(X_test)
Y_pred
```

```python
Y_test
len(Y_test)
```

```python
plt.figure(figsize=(10, 6))

# Create an index for x-axis
index = np.arange(len(Y_test))

# Plotting actual and predicted values
plt.scatter(index, Y_test, color='blue', label='Actual', alpha=0.5)
plt.scatter(index, Y_pred, color='red', label='Predicted', alpha=0.5)
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.title('Actual vs. Predicted Values')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Create a mesh grid for plotting decision boundary
X_set, y_set = X_train, Y_train
X1, X2 = np.meshgrid(np.arange(start=X_set[:, 0].min() - 1,
                               stop=X_set[:, 0].max() + 1,
                               step=0.01),
                     np.arange(start=X_set[:, 1].min() - 1,
                               stop=X_set[:, 1].max() + 1,
                               step=0.01))

# Predicting the class using the trained classifier
Z = classifier.predict(np.array([X1.ravel(), X2.ravel()]).T)
Z = Z.reshape(X1.shape)

# Plotting the decision boundary
plt.figure(figsize=(10, 6))
plt.contourf(X1, X2, Z, alpha=0.3, cmap=ListedColormap(('red', 'green')))

# Plotting the training points
plt.scatter(X_set[y_set == 0, 0], X_set[y_set == 0, 1],
            color='red', label='Not Purchased')
plt.scatter(X_set[y_set == 1, 0], X_set[y_set == 1, 1],
            color='green', label='Purchased')

# Setting labels and title
plt.title('Logistic Regression (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

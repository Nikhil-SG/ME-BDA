```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
# import openpyxl
```

```python
url = 'https://www.rottentomatoes.com/'
data = requests.get(url)
soup = BeautifulSoup(data.text, 'html.parser')
# soup
```

```python
soup.prettify()
```

```python
# Initialize an empty list to store movie data
data = []

# Loop through the carousel list and extract data
for movie in soup.find_all('div', class_='ordered-layout__list ordered-layout__list--carousel'):
    item = {}
    
    # Extract the movie title
    title_tag = movie.find('span', class_='p--small')
    item['Movie_Title'] = title_tag.text.strip() if title_tag else 'N/A'
    
    # Extract critics score
    critics_score_tag = movie.find('rt-text', slot='criticsScore')
    item['Critics_Score'] = critics_score_tag.text.strip() if critics_score_tag else 'N/A'
    
    # Extract audience score
    audience_score_tag = movie.find('rt-text', slot='audienceScore')
    item['Audience_Score'] = audience_score_tag.text.strip() if audience_score_tag else 'N/A'
    
    data.append(item)
    
    # To be polite, add a small delay between requests
    time.sleep(1)

# Convert the data to a DataFrame
df = pd.DataFrame(data)
df
# Save to Excel and CSV
# df.to_excel('rotten_tomatoes.xlsx', index=False)
# df.to_csv('rotten_tomatoes.csv', index=False)

print('Scraping Done!')
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.amazon.in/gp/bestsellers"
r=requests.get(url)
soup=BeautifulSoup(r.text,"html.parser")
```

```python
print(soup.prettify())
```

```python
ratings = soup.find_all("i")
for ratings in ratings:
    print(ratings.text)
```

```python
prices=soup.find_all("span")
for price in prices:
    print(price.text)
```

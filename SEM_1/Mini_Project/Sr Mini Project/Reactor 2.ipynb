```python
def introducing_Error(dupData):
  columns = ['Fc','Tc','Tr','Tj']

  for cols in columns:
    errors = []
    err_number = 1
    if cols == 'Fc':
      for i in range(1000):
        errors.append(0)
      while err_number <= 9000:
        err = np.random.uniform(0,orgData[cols].max()+0.4)
        if float(err) < orgData[cols].min() or float(err) > orgData[cols].max():
          if err not in errors:
            errors.append(err)
            err_number += 1
          else:
            continue
      arr1 = np.random.uniform(orgData['Tc'].min(),orgData['Tc'].max(),10000)
      arr2 = np.random.uniform(orgData['Tr'].min(),orgData['Tr'].max(),10000)
      arr3 = np.random.uniform(orgData['Tj'].min(),orgData['Tj'].max(),10000)
      if len(errors) != 0 :
        for j in range(len(errors)):
          newRow = {'Fc':errors[j],'Tc':arr1[j],'Tr':arr2[j],'Tj':arr3[j],'Target Class':1}
          dupData = dupData.append(newRow,ignore_index=True)
    else:
      while err_number <= 10000:
        err = np.random.uniform(orgData[cols].min()-5,orgData[cols].max()+5)
        if float(err) < orgData[cols].min() or float(err) > orgData[cols].max() and float(err) != 0:
          if err not in errors:
              errors.append(err)
              err_number += 1
          else:
            continue

      if cols == 'Tc':
        arr1 = np.random.uniform(orgData['Fc'].min(),orgData['Fc'].max(),10000)
        arr2 = np.random.uniform(orgData['Tr'].min(),orgData['Tr'].max(),10000)
        arr3 = np.random.uniform(orgData['Tj'].min(),orgData['Tj'].max(),10000)
        if len(errors) != 0 :
          for j in range(len(errors)):
            newRow = {'Fc':arr1[j],'Tc':errors[j],'Tr':arr2[j],'Tj':arr3[j],'Target Class':2}
            dupData = dupData.append(newRow,ignore_index=True)

      elif cols == 'Tr':
        arr1 = np.random.uniform(orgData['Tc'].min(),orgData['Tc'].max(),10000)
        arr2 = np.random.uniform(orgData['Fc'].min(),orgData['Fc'].max(),10000)
        arr3 = np.random.uniform(orgData['Tj'].min(),orgData['Tj'].max(),10000)
        if len(errors) != 0:
          for j in range(len(errors)):
            newRow = {'Fc':arr2[j],'Tc':arr1[j],'Tr':errors[j],'Tj':arr3[j],'Target Class':3}
            dupData = dupData.append(newRow,ignore_index=True)

      elif cols == 'Tj':
        arr1 = np.random.uniform(orgData['Tc'].min(),orgData['Tc'].max(),10000)
        arr2 = np.random.uniform(orgData['Tr'].min(),orgData['Tr'].max(),10000)
        arr3 = np.random.uniform(orgData['Fc'].min(),orgData['Fc'].max(),10000)
        if len(errors) != 0 :
          for j in range(len(errors)):
            newRow = {'Fc':arr3[j],'Tc':arr1[j],'Tr':arr2[j],'Tj':errors[j],'Target Class':4}
            dupData = dupData.append(newRow,ignore_index=True)

  return dupData
```

```python
X = errorData.iloc[:,[0,1,2,3]]
y = errorData.iloc[:,4]

from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(sampling_strategy='majority')
X,y = rus.fit_resample(X,y)
print("rev index:",X.shape)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf','linear','poly']}

from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(svm,param_grid)
grid.fit(X_train,y_train)
grid_svc = grid.predict(X_test)

met.accuracy_score(y_test,grid_svc)
print(grid.best_estimator_)
```

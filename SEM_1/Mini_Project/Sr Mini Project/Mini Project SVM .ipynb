# Error in Fc


```python
Fc = pd.DataFrame(data.iloc[:,1])
# len(Fc)
# Fc
```

```python
def introducing_FcError(df_Fc):
  lower_bound_index = 0
  upper_bound_index = 1000
  persistent_error_number = 10

  i = 0
  while  i < persistent_error_number:
    start_position = np.random.randint(lower_bound_index, upper_bound_index)
    end_position = start_position + 300
    errors = []
    err_number = 1
    while err_number <= 300:
      err = np.random.uniform(-1,1)
      if float(err) < 0.125 or float(err) > 0.998277:
        if err not in errors:
          errors.append(err)
          err_number += 1
      else:
        continue
    for row_num in range(start_position, end_position):
      if len(errors) != 0 :
        for j in errors:
          df_Fc = pd.DataFrame(np.insert(df_Fc.values, row_num, j, axis = 0))
          errors.pop()
          break
    lower_bound_index  = upper_bound_index
    upper_bound_index += 1300
    i += 1
  return df_Fc
```

```python
import matplotlib.pyplot as plt

def value_plot(df1, y1, df2, y2, titles=None,x_label='', y_label='', figscale=1):
    num_plots = 2
    fig, axes = plt.subplots(1, num_plots, figsize=(6 * figscale * num_plots,  3 * figscale))

    df1[y1].plot(kind='line', ax=axes[0])
    df2[y2].plot(kind='line', ax=axes[1])

    for i in range(num_plots):
        axes[i].spines[['top', 'right']].set_visible(False)
        axes[i].set_title(titles[i])
        axes[i].set_xlabel(x_label)
        axes[i].set_ylabel(y_label)

    plt.show()
```

# Creating the Target Class
---




```python
testData = batch_reactor_df
testData.shape
```

```python
def is_error_Fc(value):
  if value < correct_data_read['Fc'].min() or value > correct_data_read['Fc'].max():
    return True
  return False

def is_error_Tc(value):
  if value < correct_data_read['Tc'].min() or value > correct_data_read['Tc'].max():
    return True
  return False

def is_error_Tr(value):
  if value < correct_data_read['Tr'].min() or value > correct_data_read['Tr'].max():
    return True
  return False

def is_error_Tj(value):
  if value < correct_data_read['Tj'].min() or value > correct_data_read['Tj'].max():
    return True
  return False
```

```python
def determine_error_columns(row):
    error_columns = []
    if is_error_Fc(row['Fc']):
        error_columns.append('Fc')
    if is_error_Tc(row['Tc']):
        error_columns.append('Tc')
    if is_error_Tr(row['Tr']):
        error_columns.append('Tr')
    if is_error_Tj(row['Tj']):
        error_columns.append('Tj')
    if len(error_columns) == 0:
      return 'No Error'
    return ','.join(error_columns)

testData['Target Class'] = testData.apply(determine_error_columns, axis=1)
```
